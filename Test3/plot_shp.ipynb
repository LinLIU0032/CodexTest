# Visualize 3 SHP files (3x1 with a/b/c labels)

```python
from pathlib import Path
import geopandas as gpd
import matplotlib.pyplot as plt

# Notebook is expected to be under Test3
base_dir = Path.cwd()
shp_files = sorted(base_dir.glob('*.shp'))

if len(shp_files) != 3:
    raise ValueError(f'Expected 3 .shp files, found {len(shp_files)}: {[p.name for p in shp_files]}')

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 18))
labels = ['a', 'b', 'c']

for ax, shp_path, label in zip(axes, shp_files, labels):
    gdf = gpd.read_file(shp_path)
    gdf.plot(ax=ax, edgecolor='black', linewidth=0.6, cmap='Set3')
    ax.set_title(shp_path.name, fontsize=12)
    ax.set_axis_off()
    ax.text(0.01, 0.98, label, transform=ax.transAxes, fontsize=16, fontweight='bold', va='top', ha='left')

plt.tight_layout()
plt.show()
```
